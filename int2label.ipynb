```python
import pandas as pd
import sqlalchemy
```

```python
df = pd.read_csv('data.csv', sep=";")
initial_df = df.copy()
df.head()
```

```python
df.columns
```

```python
categorical_columns = ['Marital_status', 'Daytime_evening_attendance', 'Mothers_qualification', 
                       'Fathers_qualification', 'Mothers_occupation', 'Fathers_occupation',
                       'Displaced', 'Educational_special_needs', 'Debtor', 
                       'Tuition_fees_up_to_date', 'Gender', 'Scholarship_holder',
                       'International']
```

```python
for col in ['Mothers_qualification', 'Fathers_qualification','Mothers_occupation', 'Fathers_occupation']:
    freq = df[col].value_counts(normalize=True)
    rare_categories = freq[freq < 0.03].index
    df[col] = df[col].replace(rare_categories, "Others")
    initial_df[col] = initial_df[col].replace(rare_categories, "Others")
    print(df[col].value_counts())
    print(initial_df[col].value_counts())
```

```python
df['Status'].value_counts()
```

```python
df['Marital_status'] = df['Marital_status'].replace({1:"Single", 2:"Married", 3:"Widower", 4:"Divorced", 5:"Facto Union", 6:"Legally Separated"})
df['Daytime_evening_attendance'] = df['Daytime_evening_attendance'].replace({0:"Evening", 1:"Daytime"})
df['Mothers_qualification'] = df['Mothers_qualification'].replace({1:"Secondary Education", 3:"Higher Education", 19:"Basic Education 3rd Cycle or Equiv", 37:"Basic education 1st cycle or equiv", 38:"Basic Education 2nd Cycle or Equiv"})
df['Fathers_qualification'] = df['Fathers_qualification'].replace({37:"Basic education 1st cycle or equiv", 19:"Basic Education 3rd Cycle or Equiv", 1:"Secondary Education", 38:"Basic Education 2nd Cycle or Equiv", 3:"Higher Education - Degree"})
df['Mothers_occupation'] = df['Mothers_occupation'].replace({9:"Unskilled Workers", 4:"Administrative Staff", 5:"Personal Services, Security and Safety Workers and Sellers", 3:"Intermediate Level Technicians and Professions", 2:"Specialists in Intellectual and Scientific Activities", 7:"Skilled Workers in Industry, Construction and Craftsmen", 0:"Student"})
df['Fathers_occupation'] = df['Fathers_occupation'].replace({9:"Unskilled Workers", 7:"Skilled Workers in Industry, Construction and Craftsmen", 5:"Personal Services, Security and Safety Workers and Sellers", 4:"Administrative Staff", 3:"Intermediate Level Technicians and Professions", 8:"Installation and Machine Operators and Assembly Workers", 10:"Armed Forces Professions", 6:"Farmers and Skilled Workers in Agriculture, Fisheries and Forestry", 2:"Specialists in Intellectual and Scientific Activities", 1:"Representatives of the Legislative Power and Executive Bodies, Directors, Directors and Executive Managers"})
df['Displaced'] = df['Displaced'].replace({0:"No", 1:"Yes"})
df['Educational_special_needs'] = df['Educational_special_needs'].replace({0:"No", 1:"Yes"})
df['Debtor'] = df['Debtor'].replace({0:"No", 1:"Yes"})
df['Tuition_fees_up_to_date'] = df['Tuition_fees_up_to_date'].replace({0:"No", 1:"Yes"})
df['Gender'] = df['Gender'].replace({0:"Female", 1:"Male"})
df['Scholarship_holder'] = df['Scholarship_holder'].replace({0:"No", 1:"Yes"})
df['International'] = df['International'].replace({0:"No", 1:"Yes"})

for i in categorical_columns:
    print(df[i].value_counts())
    print("")
```
